```python
# to rename a columns 
apt.rename(columns={'type':'movie_type'},inplace=True)
apt.rename(columns={'rating':'movie_rating'},inplace=True)
apt.rename(columns={'Rating':'customer_rating'},inplace=True)
```

```python
# to change a data type from integer to string
apt['release_year'] = apt['release_year'].astype(str)
```

```python
# splitting columns containing comma delimeter
apt['main_actor']= apt['cast'].str.split (pat=',', expand=True)[0]
apt['genre']= apt['listed_in'].str.split (pat=',', expand=True)[0]
```

```python
apt.isna().sum()
```

```python
# replacing the null values in the columns 
apt['director'] = apt['director']. replace(np.NaN, 'unknown')
apt['main_actor'] = apt['main_actor']. replace(np.NaN, 'unknown')
apt['movie_rating'] = apt['movie_rating']. replace(np.NaN, 'unknown')
```

```python
# dropping irrelevant columns
apt.drop(['country','date_added','cast','listed_in','description'], axis = 1, inplace=True)
```

1. . What are the top genres available on Amazon Prime?

# VISUALIZATION

```python
def plot_genre(var, l=10, b=5):
    plt.figure(figsize=(l, b))
    sns.countplot(apt[var], order=apt[var].value_counts().index)
    plt.xticks(rotation=90)
    
plot_genre('genre')
```

 OBSERVATION

.Drama is the top genres available on Amazon Prime
.Action and comedy came second and third respectively in the top genres available on amazon prime.

# Question 3.
 How many movies are available in each release year?

```python
movies_count = apt.groupby ('release_year').size().reset_index(name= 'movie_type')
movies_count.plot(x ='release_year', y= 'movie_type', kind= 'bar',figsize = (12,6))
plt.xlabel('Release Year')
plt.ylabel('Number of movies')
plt.title('Number of movies available in each of Release Year')
plt.show()                                                                           
```

 OBSERVATION

. Year 2021 had the highest number of movies released with over 1400 movies.
. it was observed that there has been fluctuation with the number of movies released,with 2015 marking the begining of increase in the number of movies.

# Question 4.
 Who are the most prominent directors in the dataset?

```python
directors_count = apt['director'].value_counts().reset_index(name='movie_type')
directors_count.columns = ['director', 'movie_type']

plt.figure(figsize=(20, 8))
plt.bar(directors_count['director'][:10], directors_count['movie_type'][:10])
plt.xlabel('Directors')
plt.ylabel('Number of Movies')
plt.title('Most prominent Directors')
```
